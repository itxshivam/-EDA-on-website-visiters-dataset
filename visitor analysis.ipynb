●	Write a function called data_cleaning() which, when called, would perform the following activity:                    
     1.	Create a new column, called ‘Month_Year’, using lambda function. The new column should be at the 3rd position from the           start in the given dataset & its values should be : ’01-01-2020’ for January, 2020 and ’01-02-2020’ for February 2020            and so on.  (snippet added)                                                                           
     2.	Replaces the null values with the average of the respective column in the data.                                 
     3.	In column ‘B’ replace Jan with 1, feb with 2, march with 3 and so on.                                       
     4.	In column ‘E’ Replace “Came_From_LinkedIn” with “LinkedIn” and “Landed_Directly” with “Direct_traffic” .

```python
#for i in df.columns:

def data_cleaning(dataframe):
    dataframe['Month']=dataframe['Month'].astype('category')
    dataframe['Month_Year']= '01' +'-'+ dataframe['Month'].astype(str) + '-' + dataframe['Year'].astype(str)
    columns = list(dataframe.columns.values)
    dataframe= dataframe[columns[0:2] + [columns[-1]] + columns[2:-1]]
    
    dataframe['Month_Year']=pd.to_datetime(dataframe['Month_Year'])
    dataframe['Month_Year']=pd.to_datetime(dataframe['Month_Year'],format='%d-%m-%Y')
    dataframe['Month_Year']=dataframe['Month_Year'].dt.strftime('%d-%m-%Y')
    
    dataframe['Level 1'].fillna(dataframe['Level 1'].mean(),inplace=True)
    dataframe['Level 2'].fillna(dataframe['Level 2'].mean(),inplace= True)
    
    dataframe['Month']=dataframe['Month'].map({'Jan':'1',
                             'Feb':'2',
                             'Mar':'3',
                             'Apr':'4',
                             'May':'5',
                             'Jun':'6',
                             'Jul':'7',
                             'Aug':'8',
                             'Sep':'9',
                             'Oct':'10',
                             'Nov':'11',
                             'Dec':'12'
                             })
        
    dataframe['Coming from']=dataframe['Coming from'].replace('Came_From_LinkedIn','LinkedIn')
    dataframe['Coming from']=dataframe['Coming from'].replace('Landed_Directly','Direct_traffic')
        
    
    return (dataframe)
        
    
    
    
```

```python
data_cleaning(data)
```

## tasting for stationarity or not

```python
#perform Dickey-fuller test:
from statsmodels.tsa.stattools import adfuller

print('result of dickey-fuller test:')
dftest = adfuller(indexedDataset['Level 4'],autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags used','Number of observation Used'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)'%key] = value
    
print(dfoutput)
```

because (test statistic) < (critical value)  so, reject null hypothsis meaning 

Hence data is stationary
